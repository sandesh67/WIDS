```python
first_h = []
for v in lab_col_names:
    all_data[v+"_d1_value_range"] = all_data[f"d1_{v}_max"].subtract(all_data[f"d1_{v}_min"])
    
    all_data[v+"_h1_value_range"] = all_data[f"h1_{v}_max"].subtract(all_data[f"h1_{v}_min"])
#     df[v+"_h1_value_range_normalized"] = df[f"h1_{v}_max"].subtract(df[f"h1_{v}_min"]).div(df[f"h1_{v}_max"])
    
    # daily change in value range - hour vs day. could do subtract or div here.. 
    all_data[v+"_tot_change_value_range_normed"] = abs((all_data[v+"_d1_value_range"].div(all_data[v+"_h1_value_range"])))#.div(df[f"d1_{v}_max"]))
    
    # Cases where there's no reading in the first hour, but only  later in day ?
    all_data[v+"_started_after_firstHour"] = ((all_data[f"h1_{v}_max"].isna()) & (all_data[f"h1_{v}_min"].isna())) & (~all_data[f"d1_{v}_max"].isna())
    first_h.append(v+"_started_after_firstHour")
    
    ## Did a reading get more extreme after the first hour. 
    ## This misses cases where the readying is in an unhealthy region and "improves"!!
    all_data[v+"_day_more_extreme"] = ((all_data[f"d1_{v}_max"]>all_data[f"h1_{v}_max"]) | (all_data[f"d1_{v}_min"]<all_data[f"h1_{v}_min"]))
    all_data[v+"_day_more_extreme"].fillna(False)
```

```python
params = {
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'metric': 'auc',
        'learning_rate': 0.01,
        'subsample': 1,
        'colsample_bytree': 0.2,
        'reg_alpha': 3,
        'reg_lambda': 1,
        'scale_pos_weight': 1,
        'n_estimators': 30000,
        'silent': -1,
        'verbose': -1,
        'max_depth': -1
    }
```

```python
dtrain = lgb.Dataset(train, y)
evals = lgb.cv(params,
             dtrain,
             nfold=5,
             stratified=True,
             num_boost_round=20000,
             early_stopping_rounds=200,
             verbose_eval=100,
             seed = 666,
             show_stdv=True)
max(evals['auc-mean'])
```

```python
n = int(1.1 * len(evals['auc-mean']))
n
```

```python
target = y
preds = np.zeros([test.shape[0], 3])
for i in range(3):
    
    lgb_params = {
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'metric': 'auc',
        'learning_rate': 0.01,
        'subsample': 1,
        'colsample_bytree': 0.2,
        'reg_alpha': 3,
        'reg_lambda': 1,
        'scale_pos_weight': 1,
        'n_estimators': n,
        'silent': -1,
        'verbose': 200,
        'max_depth': -1,
        'seed':i + 666,
    }
    x_train = train
    x_val = train
       
    y_train = target
    y_val = target
    
    clf = lgb.LGBMClassifier(**lgb_params)
    clf.fit(
        x_train, y_train,
        eval_set=[(x_train, y_train), (x_val, y_val)],
        verbose=100,
        early_stopping_rounds=150
    )
    
    preds[:, i] = clf.predict_proba(test, num_iteration=clf.best_iteration_)[:, 1]
```

```python
y_preds = np.power(np.prod(preds, axis=1), 1/3)
```
